```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_threshold['Threshold'], df_threshold['Quality Score'], marker='o', color='blue', label='Quality Score')
plt.plot(df_threshold['Threshold'], df_threshold['Spatial Score'], marker='o', color='green', label='Spatial Score')
plt.plot(df_threshold['Threshold'], df_threshold['Quantity Score'], marker='o', color='red', label='Quantity Score')


# Annotate data points
for i, txt in enumerate(df_threshold['Quality Score']):
    plt.annotate(f'{txt:.3f}', (df_threshold['Threshold'][i], df_threshold['Quality Score'][i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(df_threshold['Spatial Score']):
    plt.annotate(f'{txt:.3f}', (df_threshold['Threshold'][i], df_threshold['Spatial Score'][i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(df_threshold['Quantity Score']):
    plt.annotate(f'{txt:.3f}', (df_threshold['Threshold'][i], df_threshold['Quantity Score'][i]), textcoords="offset points", xytext=(0,5), ha='center')


# Adding title and labels
plt.title('Threshold vs. Individual, Group, and Total')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
# plt.grid(True)

# Show plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_threshold['Threshold'], df_threshold['Average'], marker='o', color='blue', label='Average Score')

# Annotate data points
for i, txt in enumerate(df_threshold['Average']):
    plt.annotate(f'{txt:.3f}', (df_threshold['Threshold'][i], df_threshold['Average'][i]), textcoords="offset points", xytext=(0,7.5), ha='center')

# Adding title and labels
plt.title('Threshold vs. Average')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.ylim(0, 0.6)  
plt.xticks(df_threshold['Threshold'])  # Set x-axis ticks to match 'Threshold' values
# plt.grid(True)

# Show plot
plt.show() 
```
